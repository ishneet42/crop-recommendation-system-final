```python
import pandas as pd
import pickle as pkl
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#load test dataset
crop = pd.read_csv('crop_recommendation_test.csv')
crop
```

```python
crop_test = crop.iloc[:,1:-1]
crop_test
```

```python
y = crop.iloc[:,-1]
y.head
```

```python
#load the model

with open('crop_recommendation_RF.pkl', 'rb') as f:
    model = pkl.load(f)
```

```python
#predict the result
pred = model.predict(crop_test)
pred
```

```python
actual = np.array(y)
actual
```

```python
pred == actual
```

```python
#check the accuracy score
model.score(crop_test,y) 
```
